## Creating data

### DataFrame

A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.

```python
pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

### Series

A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list. And in fact you can create one with nothing more than a list:

```python
import numpy as np
pd.Series([1, 2, 3, np.nan, 4, 5])
```

**Note:** a Series does not have a column name, it only has one overall name

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

**Note:** It's helpful to think of a DataFrame as actually being just a bunch of Series $\textit{glued together}$.

```python
animals = pd.DataFrame({'Cows': [12, 20], 'Goats': [22, 19]}, index=['Year 1', 'Year 2'])
# save this DataFrame to disk as a csv file with the name cows_and_goats.csv.
animals.to_csv("cows_and_goats.csv")
# save this DataFrame to disk as an excel file with the name cows_and_goats.xlsx.
animals.to_excel("cows_and_goats.xlsx")

animals
```

## Maps

A **map** is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later.

There are two mapping methods that you will use often.

$\textit{map()}$ is the first, and slightly simpler one. $\textit{map()}$ returns a new Series where all the values have been transformed by your function. For example, suppose that we wanted to remean the scores the wines received to $0$. We can do this as follows:

```python
review_points_mean = wine_reviews.points.mean()
wine_reviews.points.map(lambda p: p - review_points_mean)
```

**Note:** Pandas provides many common mapping operations as built-ins. For example, here's a faster way of remeaning our points column:

```python
review_points_mean = wine_reviews.points.mean()
wine_reviews.points - review_points_mean
```

Pandas will also understand what to do if we perform these operations between Series of equal length. For example, an easy way of combining country and region information in the dataset would be to do the following:

```python
wine_reviews.country + " - " + wine_reviews.region_1
```

These operators are faster than $\textit{map()}$ or $\textit{apply()}$ because they uses speed ups built into pandas. All of the standard Python operators (>, <, ==, and so on) work in this manner.

However, they are not as flexible as $\textit{map()}$ or $\textit{apply()}$, which can do more advanced things, like applying conditional logic, which cannot be done with addition and subtraction alone.

$\textit{apply()}$ is the equivalent method if we want to transform a whole DataFrame by calling a custom method on each $\underline{row}$.

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

wine_reviews.apply(remean_points, axis='columns')
```

If we had called $\textit{wine_reviews.apply()}$ with $\textit{axis='index'}$, then instead of passing a function to transform each row, we would need to give a function to transform each column.

**Note:** $\textit{map()}$ and $\textit{apply()}$ return new, transformed Series and DataFrames, respectively. They don't modify the original data they're called on. (Copy or View?)

There are only so many words you can use when describing a bottle of wine. Is a wine more likely to be $\textit{"tropical"}$ or $\textit{"fruity"}$? We want to create a Series descriptor_counts counting how many times each of $\textit{'tropical'}$ and $\textit{'fruity'}$ appears in the description column in the dataset. we can do this, like the following:

```python
n_trop = wine_reviews.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = wine_reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])

print(descriptor_counts)
```

We'd like to host these wine reviews on our website, but a rating system ranging from $80$ to $100$ points is too hard to understand - we'd like to translate them into simple star ratings. A score of $95$ or higher counts as $3$ stars, a score of at least $85$ but less than $95$ is $2$ stars. Any other score is $1$ star.

Also, the Canadian Vintners Association bought a lot of ads on the site, so any wines from Canada should automatically get $3$ stars, regardless of points.

Create a series star_ratings with the number of stars corresponding to each review in the dataset.

```python
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star_ratings = wine_reviews.apply(stars, axis='columns')
print(star_ratings)
```

## Indexing in pandas

### Index-based selection

To select the first row of data in a DataFrame, we may use the following:

```python
wine_reviews.iloc[0]
```

```python
wine_reviews.price.iloc[1] # equals to wine_reviews['price'][1]. but first one is preferred
```

To get a column with iloc, we can do the following:

```python
wine_reviews.iloc[:, 0] # the : operator, which also comes from native Python, means "everything".
```

It's also possible to pass a list:

```python
wine_reviews.iloc[[0, 1, 2], 0]
```

here are the last five elements of the dataset:

```python
wine_reviews.iloc[-5:]
```

The second paradigm for attribute selection is the one followed by the $\textit{loc}$ operator: label-based selection. In this paradigm, it's the data index value, not its position, which matters.

For example, to get the first entry in wine_reviews, we would now do the following:

```python
wine_reviews.loc[0, 'country']
```

```python
wine_reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

If we want to select the country, province, region_1, and region_2 columns of the records with the index labels 0, 1, 10, and 100, we can use the following code:

```python
cols = ['country', 'province', 'region_1', 'region_2']
indices = [0, 1, 10, 100]
wine_reviews.loc[indices, cols]
```

**Note:** iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10. This is particularly confusing when the DataFrame index is a simple numerical list, e.g. 0,...,1000. In this case df.iloc[0:1000] will return 1000 entries, while df.loc[0:1000] return 1001 of them! To get 1000 elements using loc, you will need to go one lower and ask for df.loc[0:999].

**Note:** If we have a DataFrame with index values Apples, ..., Potatoes, ..., and we want to select "all the alphabetical fruit choices between Apples and Potatoes", then it's a lot more convenient to index df.loc['Apples':'Potatoes'] than it is to index something like df.loc['Apples','Potatoes'] (P coming after A in the alphabet).

```python
wine_reviews.loc[:, 'points':'taster_name']
```

**Note:** Both $\textit{loc}$ and $\textit{iloc}$ are $\textit{row-first}$, $\textit{column-second}$. <u>This is the opposite of what we do in native Python</u>, which is column-first, row-second.

## Manipulating the index

Label-based selection derives its power from the labels in the index. The index we use is not immutable. We can manipulate the index in any way we see fit. The $\textit{set_index()}$ method can be used to do the job.

```python
wine_reviews.set_index("title")
```

This is useful if you can come up with an index for the dataset which is better than the current one.

## Conditional selection

Suppose that we're interested specifically in better-than-average wines produced in Italy. We can start by checking if each wine is Italian or not:

```python
wine_reviews.country == 'Italy'
```

This result can then be used inside of loc to select the relevant data:

```python
wine_reviews.loc[wine_reviews.country == 'Italy']
```

We also wanted to know which ones are better than average. Wines are reviewed on a 80-to-100 point scale.

```python
print(min(wine_reviews.points))
print(max(wine_reviews.points))
```

So this could mean wines that accrued at least $90$ points. We can use the ampersand ($\textit{&}$) to bring the two questions together:

```python
wine_reviews.loc[(wine_reviews.country == 'Italy') & (wine_reviews.points >= 90)]
```

Suppose we'll buy any wine that's made in Italy $\underline{or}$ which is rated above average. For this we use a pipe ($\textit{|}$):

```python
wine_reviews.loc[(wine_reviews.country == 'Italy') | (wine_reviews.points >= 90)]
```

$\textit{isin}$ is lets you select data whose value $\textit{"is in"}$ a list of values. For example, here's how we can use it to select wines only from Italy or France:

```python
wine_reviews.loc[wine_reviews.country.isin(['Italy', 'France'])]
```

$\textit{isnull}$ (and $\textit{notnull}$) methods let you highlight values which are (or are not) empty (NaN). For example, to filter out wines lacking a price tag in the dataset, here's what we would do:

```python
wine_reviews.loc[wine_reviews.price.notnull()]
```

## Missing Data

**Note:** Entries missing values are given the value $\textit{NaN}$, short for $\textit{"Not a Number"}$. For technical reasons these NaN values are always of the <u>float64</u> dtype.

To select NaN entries you can use $\textit{pd.isnull()}$ (or its companion $\textit{pd.notnull()}$).

```python
wine_reviews[pd.isnull(wine_reviews.country)]
```

```python
wine_reviews[pd.notnull(wine_reviews.country)]
```

Replacing missing values is a common operation. $\textit{fillna()}$ provides a few different strategies for mitigating such data.

```python
wine_reviews.region_2.fillna("Unknown") # replacing each NaN entries with "Unknown", in region_2 column
```

Create a Series counting the number of times each value occurs in the region_1 field. This field is often missing data, so replace missing values with Unknown. Sort in descending order.

```python
wine_reviews.region_1.fillna("Unknown").value_counts().sort_values(ascending=False)
```

Alternatively, we may have a non-null value that we would like to replace. For example, suppose that since this dataset was published, reviewer Kerin O'Keefe has changed her Twitter handle from @kerinokeefe to @kerino. One way to reflect this in the dataset is using the $\textit{replace()}$ method:

```python
wine_reviews.taster_twitter_handle.replace("@kerinokeefe", "@kerino")
```

## Groupwise analysis

```python
wine_reviews.groupby('points')
```

```python
wine_reviews.groupby('points').count()
```

One function we've been using heavily thus far is the $\textit{value_counts()}$ function.

```python
wine_reviews.points.value_counts()
```

We can replicate what value_counts() does by doing the following:

```python
wine_reviews.groupby('points').points.count()
```

$\textit{value_counts()}$ is just a shortcut to this $\textit{groupby()}$ operation.

We can use any of the summary functions we've used before with this data. For example, to get the cheapest wine in each point value category, we can do the following:

```python
wine_reviews.groupby('points').price.min()
```

You can think of each group we generate as being a slice of our DataFrame containing only data with values that match. Here's one way of selecting the name of the first wine reviewed from each winery in the dataset:

```python
wine_reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
for i in range(len(wine_reviews['winery'])):
    if wine_reviews.winery.iloc[i] == '1+1=3':
        print(wine_reviews.title[i])
```

Here's how we would pick out the best wine by country and province:

```python
wine_reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

method worth mentioning is $\textit{agg()}$, which lets you run a bunch of different functions on your DataFrame simultaneously. For example, we can generate a simple statistical summary of the dataset as follows:

```python
wine_reviews.groupby(['country']).price.agg([len, min, max])
```

```python
countries_reviewed = wine_reviews.groupby(['country', 'province']).price.agg([len, min, max])
countries_reviewed
```

In general, the multi-index method you will use most often is the one for converting back to a regular index, the $\textit{reset_index()}$ method:

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed
```

## Renaming

$\textit{rename()}$ method, lets us change index names and/or column names.

```python
wine_reviews.rename(columns={'points': 'score'}) # change the points column in our dataset to score
```

$\textit{rename()}$ lets you rename index or column values by specifying a index or column keyword parameter, respectively.  It supports a variety of input formats, but usually a Python dictionary is the most convenient.

```python
wine_reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

$\textbf{Note:}$ You'll probably rename columns very often, but rename index values very rarely. For that, $\textit{set_index()}$ is usually more convenient.

Both the row index and the column index can have their own name attribute. The complimentary $\textit{rename_axis()}$ method may be used to change these names. For example:

```python
wine_reviews.rename_axis("wines", axis='rows')
```

```python
wine_reviews.rename_axis("fields", axis='columns')
```
